```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset
df = pd.read_csv('auto-mpg.csv')
df
```

```python
for i in range(len(df['horsepower'])):
  if df['horsepower'][i] == '?':
    df['horsepower'][i] = 0
    print(i)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['horsepower'] = df['horsepower'].astype('int64')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
# Select the input features and output target variable
X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin']]
X
```

```python
X.dtypes
```

```python
Y = df['mpg']
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```

```python
# evaluating model's generalization ability
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
